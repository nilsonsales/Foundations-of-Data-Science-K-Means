```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("bank-notes.csv")
```

```python
data
```

```python
class_1 = data[ data["Class"] == 1 ]
class_2 = data[ data["Class"] == 2]
```

```python
# Analysing statistical metrics for the classes:

mean_1 = [ class_1[column].mean() for column in class_1.columns ]
mean_2 = [ class_2[column].mean() for column in class_2.columns ]

print("Class 1 means: ", mean_1)
print("Class 2 means: ", mean_2)
```

```python
std_1 = [ class_1[column].std() for column in class_1.columns ]
std_2 = [ class_2[column].std() for column in class_2.columns ]

print("Class 1 std dev: ", std_1)
print("Class 2 std dev: ", std_2)
```

```python
# Visualising the data

plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(class_1['V1'], class_1['V2'], label="Class 1", alpha=0.5)
plt.scatter(class_2['V1'], class_2['V2'], label="Class 2", alpha=0.5)
plt.text(5,12,"Size = V3")


plt.title("Figure 1")
plt.legend()
plt.show()
```

```python
plt.xlabel('V1')
plt.ylabel('V4')
plt.scatter(class_1['V1'], class_1['V4'], s=class_1['V3']*2, label="Class 1", alpha=0.5)
plt.scatter(class_2['V1'], class_2['V4'], s=class_2['V3']*2, label="Class 2", alpha=0.5)
plt.text(5, -8.5,"Size = V3")

plt.title("Fugure 2")
plt.legend()
plt.show()
```

```python
plt.xlabel('V2')
plt.ylabel('V3')
plt.scatter(class_1['V2'], class_1['V3'], s=9, label="Class 1", alpha=0.5)
plt.scatter(class_2['V2'], class_2['V3'], s=9, label="Class 2", alpha=0.5)


plt.title("Figure 3")
plt.legend()
plt.show()
```

```python
plt.xlabel('V2')
plt.ylabel('V4')
plt.scatter(class_1['V2'], class_1['V4'], s=9, label="Class 1", alpha=0.5)
plt.scatter(class_2['V2'], class_2['V4'], s=9, label="Class 2", alpha=0.5)


plt.legend()
plt.show()
```
